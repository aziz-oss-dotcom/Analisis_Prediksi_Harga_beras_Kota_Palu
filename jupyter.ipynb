```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Membaca dataset
data = pd.read_csv("harga_beras_palu.csv")

print("--- 5 Data Pertama ---")
data.head()
```

```python
# Mengatur ukuran kanvas grafik
plt.figure(figsize=(12, 6))  
plt.plot(data["bulan"], data["Harga_Beras"], marker='o', color='b', markersize=4, linewidth=1.5)

# Merapikan sumbu X agar label tanggal melompat tiap 6 bulan
lokasi_ticks = range(0, len(data["bulan"]), 6)
label_ticks = [data["bulan"][i] for i in lokasi_ticks]
plt.xticks(lokasi_ticks, label_ticks, rotation=30, ha='right', fontsize=9) 

# Atribut Grafik
plt.title("Analisis Prediksi Harga Beras Kota Palu", fontsize=14, fontweight='bold', pad=15)
plt.xlabel("Periode (Bulan)", fontsize=11, labelpad=10)
plt.ylabel("Harga (Rp / Kg)", fontsize=11, labelpad=10)
plt.grid(True, linestyle='--', alpha=0.5) 

plt.tight_layout()
plt.show()
```

```python
# Menentukan Fitur (X) dan Target (y)
X = data[["Inflasi", "Curah_Hujan"]]
y = data["Harga_Beras"]

# Inisialisasi dan Training Model
model = LinearRegression()
model.fit(X, y)

print("Model Regresi Linear Berhasil Dilatih!")
print(f"Intercept (Konstanta): {model.intercept_}")
print(f"Coefficients (Koefisien Fitur): {model.coef_}")
```

```python
# Contoh input: Inflasi 0.50% dan Curah Hujan 120 mm
inflasi_input = 0.50
curah_hujan_input = 120

prediksi = model.predict([[inflasi_input, curah_hujan_input]])

print(f"Hasil Prediksi Harga Beras: Rp {prediksi[0]:.2f}")
```
